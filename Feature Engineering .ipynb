```python
#Imports and Datasets
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Let's load our data
df = pd.read_csv("insurance_claims.csv")
```

```python
df.head()
```

```python
#checking missing values
null_counts = df.isnull().sum()
null_counts
```

```python
df_new = df.drop("_c39", axis = 1)
```

```python
df_new.head()
```

```python
#Binding numeric Data
df_new.describe()
```

```python
# Choose the column for the histogram
column_name = 'age'

# Plot the histogram
plt.hist(df[column_name], bins=3, edgecolor='black')

# Add labels and title
plt.xlabel(column_name)
plt.ylabel('Frequency')
plt.title(f'Histogram of {column_name}')

# Display the histogram
plt.show()
```

```python
bin_edges = [0, 30, 55, 100]  # Define the bin edges
bin_labels = ['Young Adult', 'Middle Aged', 'Elderly']  # Corresponding labels for each bin

# Create a new column based on the bin labels
df_new['ages_category'] = pd.cut(df_new['age'], bins=bin_edges, labels=bin_labels)
```

```python
bin_edges_customer = [0, 25, 150, 500]  # Define the bin edges
bin_labels_customer = ['New Client', 'Established Client', 'Long-Term Client']  # Corresponding labels for each bin

# Create a new column based on the bin labels
df_new['customer_category'] = pd.cut(df_new['months_as_customer'], bins=bin_edges_customer, labels=bin_labels_customer)
```

```python
bin_edges_vehicle = [0, 2004, 2009, 2015, ]  # Define the bin edges
bin_labels_vehicle = ['Old Vehicle', 'Semi-Old Vehicle', 'New Vehicle']  # Corresponding labels for each bin

# Create a new column based on the bin labels
df_new['Vehicle_Age_Category'] = pd.cut(df_new['auto_year'], bins=bin_edges_vehicle, labels=bin_labels_vehicle)
```

```python
df_new.head()
```

```python
bin_edges_Annual_Premium = [0,500 , 1400, 1700]  # Define the bin edges
bin_labels_Annual_Premium = ['Low-Premium', 'Mid-Leve-Premiuml', 'High-Premium']  # Corresponding labels for each bin

# Create a new column based on the bin labels
df_new['Premium_category'] = pd.cut(df_new['policy_annual_premium'], bins=bin_edges_Annual_Premium, labels=bin_labels_Annual_Premium)
```

```python
df_new.head()
```

```python
#Creating New Features
df_new["Monthly_Premium"] = df_new["policy_annual_premium"]/12
```

```python
#Feature joining
df_new['Number_of_Premiums_Paid'] = (df_new['policy_bind_date']) / df_new['months_as_customer']
```

```python
df_new.head()
```

```python
df_new['total_premiums_paid'] = (df_new['policy_annual_premium']/12) * df_new['months_as_customer']
```

```python
df_new.head()
```

```python
df_new['net_value_of_customer'] = df_new['total_premiums_paid'] - df_new['total_claim_amount']
```

```python
df_new.head()
```

```python
df_new.to_csv('Advanced Feature Claims Data.csv')
```
